```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import ndimage
```

```python
img1=cv2.imread('/Users/shuvechchhasanyal/Downloads/p1.png')
img1.shape
plt.imshow(img1)
```

```python
img2=cv2.imread('/Users/shuvechchhasanyal/Downloads/p2.png')
img2.shape
plt.imshow(img2)
```

```python
img3=cv2.imread('/Users/shuvechchhasanyal/Downloads/p3.png')
img3.shape
plt.imshow(img3)
```

```python
gray1 = rgb2gray(img1)
plt.imshow(gray1, cmap='gray')
```

```python
gray2 = rgb2gray(img2)
plt.imshow(gray2, cmap='gray')
```

```python
gray3 = rgb2gray(img3)
plt.imshow(gray3, cmap='gray')
```

```python
gray_r1 = gray1.reshape(gray1.shape[0]*gray1.shape[1])
for i in range(gray_r1.shape[0]):
    if gray_r1[i] > gray_r1.mean():
        gray_r1[i] = 1
    else:
        gray_r1[i] = 0
gray1 = gray_r1.reshape(gray1.shape[0],gray1.shape[1])
plt.imshow(gray1, cmap='gray')
```
